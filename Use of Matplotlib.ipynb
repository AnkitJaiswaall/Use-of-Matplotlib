```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# 2nd method for ploting which is Object oriented method

fig=plt.figure()
axes=fig.add_axes(0.1,0.1,0.8.0.8)

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(x,y)

axes1.set_title('Large Title1')
axes2.set_title('smaller Title')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First plot')
axes[1].plot(y,x)
axes[1].set_title('second plot')
```

```python
# fig size and DPI

fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[0].set_title('First plot')
axes[1].plot(y,x)
axes[1].set_title('second plot')

plt.tight_layout()
```

```python
fig.savefig('FirstPlot.png',dpi=200)
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x**2,label='x Square')
ax.plot(x**3,label='x cube')
ax.legend()
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='Green',lw=3,ls='-.',marker='o')
```
